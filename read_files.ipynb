# SELECT COLUMNS OR ROWS WİTH PANDAS 


```python
#select column
print(df["price"])
```

```python
#select column with the "loc" method
# loc Method = df.loc[rows,"columns"]
result = df.loc[0,"price"]
print(result)
```

```python
# Just select the rows => df.loc["rows"]
result=df.loc[0]
print(result)
```

```python
# Just select the columns => df.loc[:,["column1","column2".....]] or df.loc[:,"column1"]
result = df.loc[:,"price"]
print(result)
```

```python
# we can make the normal slicing processes
result = df.loc[:100,["price"]]
print(result)
```

```python
# returns the rows from "0" to "10" and the columns from begining to "price"
result = df.loc[:10,:"price"]
print(result)
```

```python
# you can be returned whatever you want value 
result=df.loc[0,["price"]] # it involves the last value 
print(result)
```

```python
# use to the index numbers with "iloc" method
result = df.iloc[:10,:5] # it doesn't involve the last value
print(result)
```

**-> head() and tail() methods**
- our data can be so heavy or big, therefore to take all data always cannot be true. 

- in this situation, we must have the slicing, selecting process, we can use those methods in pandas dataframes

```python
data = df.loc[:10,:"bathrooms"]
data["kitchen"]=pd.Series(np.random.randint(10,30,11),index=[i for i in range(11)])
data["locate"] =pd.Series([i for i in np.random.randint(1,100,11)],dtype=float)
print(data)
```

```python
print(data.columns) # we can return the columns name into a list
```

```python
result=data.head() # we can reach first 5 datas by using the head() method
print(result)
```

```python
# head() method can take a optional value. When it is used alone, it returns first 5 value but you can specify you want to how many datas.
result=data.head(7)
print(result)
```

**-> tail() method**
- it is the same with head() method but it starts to take last 5 methods

```python
result=data.tail()
print(result) 
```

```python
result=data.tail(10)
print(result)
```

#DELETİON PROCESS
-**drop() method** is used to delete any column or row

***Usage:***
- drop(data,axis=(int),inplace=True/False)
- if the inplace value is True, it affects to real data
- but if it is the False, it isn't affect to real data


```python
print(f"Data:\n{data}")
```

```python
result=data.drop('ANY',axis=1,inplace=False)
print(result)
print(data)
```

```python
data.drop('ANY',axis=1,inplace=True)
print(data)
```
