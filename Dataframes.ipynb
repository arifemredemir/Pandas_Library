**Dataframe** is formed by combining 2 series

```python
import numpy as np
s1=pd.Series([1,2,3,4])
s2=pd.Series(["apple","oranges","berry","oil"])

data = dict(numbers=s1,fruits=s2)

df=pd.DataFrame(data,index=[i for i in range(1,5)])

df["cars"]=pd.Series(np.random.randint(0,10,4),index=[i for i in range(1,5)],dtype=float)

print(df)
```

```python
data = [["ali",60],["mehmet",70],["ahmet",35],["yağmur",40]]
df = pd.DataFrame(data)
print(df)
```

```python
df=pd.DataFrame(data,columns=["Names","Grades"])
print(df)
```

```python
df=pd.DataFrame(data,columns=["Names","Grades"],index=[1,2,3,4],dtype=float)
print(df)
```

```python
# you can determine the index numbers whatever you want 
df=pd.DataFrame(data,columns=["Names","Grades"],index=[165,2221,321,457],dtype=float)
print(df)
```

```python
data = {"Names":["Ali","Mehmet","Halil","Yağmur"],"Grades":["25","60","90","15"]}
df=pd.DataFrame(data,dtype=float)
print(df)
```
